## 1. Principales Estadisticas de las variables

La información de los datos es la siguiente:

- fechaAlquiler: hora por hora + marca de tiempo

- codEstacion:
    - 1 = primavera
    - 2 = verano
    - 3 = caída
    - 4 = invierno

- flgFeriado: si el día se considera un día festivo

- flgDiaLaborable: si el día no es un fin de semana ni un feriado.

- codClima:

    - 1: Despejado, pocas nubes, medio nublado, parcialmente nublado
    - 2: Niebla + Nublado, Niebla + parcialmente nublado, Niebla + Pocas nubes, Niebla
    - 3: Nieve ligera, Lluvia ligera + Tormenta eléctrica + Nubes dispersas, Lluvia ligera + Nubes dispersas
    - 4: Lluvia intensa + Paletas de hielo + Tormenta + Niebla, Nieve + Niebla


- temperatura: temperatura en grados Celsius

- sensacionTermica: "sensación térmica" la temperatura en grados Celsius

- humedad: humedad relativa

- velViento: velocidad del viento

- NroAlquilerNoSocio: número de alquileres de usuarios no registrados iniciados

- NroAlquilerSocio: número de alquileres de usuarios registrados iniciados

- NroAlquilerTotal: número de alquileres totales

```python
dfBicicletas.groupby(['codEstacion']).agg(sum_total_alquiler=('NroAlquilerTotal','sum'),
                                          count_total_alquiler = ('NroAlquilerTotal', 'count'),
                                          mean_total_alquiler = ('NroAlquilerTotal', 'mean'))
```

```python
dfBicicletas['DesEstacion'] = dfBicicletas['codEstacion'].map({1:'Primavera',
                                                               2:'Verano',
                                                               3:'Otoño',
                                                               4:'Invierno'})


dfBicicletas['DesClima'] = dfBicicletas['codClima'].map({1:'Despejado',
                                                         2:'Niebla',
                                                         3:'Nieve',
                                                         4:'Llueve'})
```

```python
dfBicicletas.head()
```

```python
dfBicicletas.groupby(['DesEstacion','flgFeriado']).agg(sum_total_alquiler=('NroAlquilerTotal','sum'),
                                          count_total_alquiler = ('NroAlquilerTotal', 'count'),
                                          mean_total_alquiler = ('NroAlquilerTotal', 'mean'))
```

```python
dfBicicletas['DesFeriado'] = dfBicicletas['flgFeriado'].map({1:'Feriado',
                                                             0:'No feriado'})

dfBicicletas['DesLaboral'] = dfBicicletas['flgDiaLaborable'].map({1:'Se trabaja',
                                                             0:'No se trabaja'})
```

```python
dfBicicletas.groupby(['DesLaboral','DesFeriado']).agg(sum_total_alquiler=('NroAlquilerTotal','sum'),
                                          count_total_alquiler = ('NroAlquilerTotal', 'count'),
                                          mean_total_alquiler = ('NroAlquilerTotal', 'mean'))
```

```python
dfBicicletas.groupby(['DesFeriado','DesEstacion']).agg(sum_total_alquiler=('NroAlquilerTotal','sum'),
                                          count_total_alquiler = ('NroAlquilerTotal', 'count'),
                                          mean_total_alquiler = ('NroAlquilerTotal', 'mean'))
```

Puedes aplicar diversas funciones estadísticas y de agregación a estos grupos, como sum(), mean(), min(), max(), count(), std(), var(), etc. Puedes también aplicar funciones personalizadas usando el método apply().

```python
dfBicicletas.groupby('DesFeriado').agg(prom_alq=('NroAlquilerTotal','mean'), \
                          prom_alq_mem=('NroAlquilerSocio','mean'), \
                          prom_alq_sin = ('NroAlquilerNoSocio','mean'), \
                          prom_temp = ('temperatura','mean'))
```

```python
dfAgg = dfBicicletas.groupby('flgFeriado').agg(prom_temperatura=('temperatura','mean'), \
                          var_temperatura=('temperatura','var'), \
                          dsv_temperatura = ('temperatura','std'))
```

```python
# Coeficiente de variación
dfAgg['coeficiente_variacion'] = (dfAgg['dsv_temperatura'] / dfAgg['prom_temperatura']) * 100
dfAgg['coeficiente_variacion']
```

```python
dfAgg
```

La interpretación del coeficiente de variación es la siguiente:

* Si el CV es bajo (por ejemplo, menor al 10%), se considera que la serie de datos tiene poca variación relativa. Esto significa que los valores individuales están bastante cerca de la media y la dispersión entre ellos es relativamente pequeña.
* Si el CV es moderado (por ejemplo, entre el 10% y el 30%), se considera que la serie de datos tiene una variación moderada. Esto implica que los valores individuales se encuentran moderadamente dispersos alrededor de la media.
* Si el CV es alto (por ejemplo, mayor al 30%), se considera que la serie de datos tiene una alta variación relativa. Esto indica que los valores individuales están ampliamente dispersos y distantes de la media.
