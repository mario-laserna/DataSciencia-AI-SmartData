**Pregunta 6**

Principal uso de la Librería Pandas Dataframe:



```python
# prompt: Cual es el principal uso de la libreria Pandas Dataframe

El principal uso de la librería Pandas DataFrame es proporcionar una estructura de datos tabular bidimensional, similar a una hoja de cálculo o una tabla SQL.  Permite manipular, analizar y limpiar datos de manera eficiente.  Sus principales usos incluyen:

* **Almacenamiento y manipulación de datos:** Organizar datos en filas (observaciones) y columnas (variables).
* **Limpieza y preprocesamiento de datos:** Manejar valores faltantes, duplicados y realizar transformaciones.
* **Análisis exploratorio de datos:** Calcular estadísticas descriptivas, realizar agrupaciones y visualizar datos.
* **Preparación de datos para el modelado:** Transformar y preparar los datos para usarlos en modelos de machine learning.
* **Análisis de series temporales:**  Trabajar con datos indexados por tiempo.


En resumen, Pandas DataFrame es una herramienta fundamental para la ciencia de datos y el análisis de datos en Python.
```

**Pregunta 9**

La función MERGE de la librería Pandas sirve para:

```python
# prompt: La función MERGE de la librería Pandas sirve para:

La función `merge` de la librería Pandas sirve para combinar dos o más DataFrames en uno solo, basándose en una o varias columnas comunes entre ellos.  Esencialmente, realiza uniones similares a las que se hacen en SQL.  Permite diferentes tipos de unión (inner, outer, left, right) para controlar qué filas se incluyen en el resultado final, dependiendo de la presencia o ausencia de valores coincidentes en las columnas de unión.
```

**Pregunta 10**

El método APPLY en Pandas Dataframe se utiliza para:

```python
# prompt: El método APPLY en Pandas Dataframe se utiliza para:

El método `apply` en Pandas DataFrame se utiliza para aplicar una función a lo largo de un eje del DataFrame (ya sea filas o columnas). Permite realizar operaciones personalizadas en cada fila o columna, o en un subconjunto de ellas.  Es una herramienta muy versátil para realizar transformaciones y cálculos más complejos que los que se pueden lograr con funciones integradas.
```
