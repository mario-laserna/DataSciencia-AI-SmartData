Ejercicio 4
Escribir un programa que almacene las asignaturas de un curso (por ejemplo Matemáticas, Física, Química, Historia y Lengua) en una lista, pregunte al usuario la nota que ha sacado en cada asignatura y elimine de la lista las asignaturas aprobadas. Al final el programa debe mostrar por pantalla las asignaturas que el usuario tiene que repetir.

```python
#opción 1: copiar el diccionario en otra variable por que si se hace sobre el mismo en el for arroja un error,
#esta nueva variable finalmente va a contener solo las materias reprobadas

asignaturas_con_nota = {'Matemáticas': 2, 'Física': 3, 'Química': 4.5, 'Historia': 1.8, 'Lengua': 3.1}
asignaturas_reprobadas = asignaturas_con_nota.copy()

for asignatura in asignaturas_con_nota:
  if asignaturas_con_nota[asignatura] >= 3:
    del asignaturas_reprobadas[asignatura]

asignaturas_reprobadas
```

```python
#opción 1: usar la función copy() del diccionario en el for para estar iterando sobre una copia que no se modifica en el for

asignaturas_con_nota = {'Matemáticas': 2, 'Física': 3, 'Química': 4.5, 'Historia': 1.8, 'Lengua': 3.1}

for asignatura in asignaturas_con_nota.copy():
  if asignaturas_con_nota[asignatura] >= 3:
    del asignaturas_con_nota[asignatura]

asignaturas_con_nota
```

```python
#opción 3: se itera sobre una lista de claves y no sobre el diccionario

asignaturas_con_nota = {'Matemáticas': 2, 'Física': 3, 'Química': 4.5, 'Historia': 1.8, 'Lengua': 3.1}

# Iteramos sobre una copia de las claves para modificar el diccionario original
for asignatura in list(asignaturas_con_nota):
    if asignaturas_con_nota[asignatura] >= 3:
        del asignaturas_con_nota[asignatura]

# Imprimimos las asignaturas reprobadas (las que quedan en el diccionario)
asignaturas_con_nota
```
