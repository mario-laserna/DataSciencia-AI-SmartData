# Definición del proyecto

**Fuga de Clientes de Telecomunicaciones**<br>
Contexto: Actualmente se vienen manifestando fuga de clientes, el cual se encuentra por encima del mercado (15%), por lo que es necesario conocer el comportamiento que los motivo a retirarse de la compañía a través de la aplicación de análisis de datos y predecir el comportamiento para retener a los clientes a través de algoritmos de Machine Learning.


## 0. Objetivo
Desarrollar programas específicos de retención de clientes

###Notas Adicionales
El conjunto de datos incluye información sobre:
<ol><li>
Clientes que se fueron en el último mes: la columna se llama Churn<li>
Servicios a los que cada cliente se ha suscrito: teléfono, líneas múltiples, Internet, seguridad en línea, respaldo en línea, protección de dispositivos, soporte técnico y transmisión de TV y películas. <li>Información de la cuenta del cliente: cuánto tiempo ha sido cliente, contrato, método de pago, facturación electrónica, cargos mensuales y cargos totales<li>
Información demográfica sobre los clientes: género, rango de edad y si tienen parejas y dependientes</ol>

##8. Balanceo de datos

```python
# Seleccionaremos un 30% de los datos para estrudiar el beneficio de las estretgias de muestreo
#Variables independientes
X = fuga_encoded[['tenure', 'PaymentMethod_encoded', 'Contract_encoded', 'InternetService_encoded']] #Se colocan las variables + importantes
#Variable dependiente
y = fuga_encoded[['Churn']]

import sklearn.model_selection as model_selection
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, train_size=0.70,test_size=0.30, random_state=1206)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
# Solo utilizaremos Datos de entrenamiento para la estrategia
df_train = pd.concat([X_train,y_train],axis=1)
df_train.head()
```

```python
count_class_0, count_class_1 = df_train.Churn.value_counts()

# Divide by class
# se parte nuevamente en dos bases
df_class_0 = df_train[df_train['Churn'] == 0]
df_class_1 = df_train[df_train['Churn'] == 1]
```

```python
count_class_0, len(df_class_0), count_class_1, len(df_class_1)
```

```python
3635+1295
```

```python
1295/4930
```

###8.1 Undersampling

```python
df_class_0_under = df_class_0.sample(count_class_1,random_state=1206)
```

```python
df_train_under = pd.concat([df_class_0_under, df_class_1], axis=0)
```

```python
print('Estrategia de Muestreo under-sampling:')
print(df_train_under.Churn.value_counts())

df_train_under.Churn.value_counts().plot(kind='bar', title='Cantidad (target)');
```

###8.2 Oversampling

```python
df_class_1_over = df_class_1.sample(count_class_0, replace=True)
```

```python
df_train_over = pd.concat([df_class_0, df_class_1_over], axis=0)
```

```python
print('Estrategia de Muestreo over-sampling:')
print(df_train_over.Churn.value_counts())

df_train_over.Churn.value_counts().plot(kind='bar', title='Cantidad (target)');
```

```python
#Train Original
df_train.Churn.value_counts()
```

```python
#Train Undersampling
df_train_under.Churn.value_counts()
```

```python
#Train Oversampling
df_train_over.Churn.value_counts()
```

###8.3 Evaluación del Balanceo

```python
#Original
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, log_loss

#Declarando el modelo
model = DecisionTreeClassifier()
#Entrenando el modelo
model.fit(X_train, y_train)
#Prediciendo el modelo
y_pred_1 = model.predict(X_test)
```

```python
#Undersampling

model = DecisionTreeClassifier()
#Defino mi base de train
X_train_u= df_train_under[['tenure', 'PaymentMethod_encoded', 'Contract_encoded', 'InternetService_encoded']]
y_train_u = df_train_under[['Churn']]
#Entrenando el modelo
model.fit(X_train_u, y_train_u)
#Prediciendo el modelo
y_pred_2 = model.predict(X_test)

```

```python
#Oversampling

model = DecisionTreeClassifier()
#Defino mi base de train
X_train_o= df_train_over[['tenure', 'PaymentMethod_encoded', 'Contract_encoded', 'InternetService_encoded']]
y_train_o = df_train_over[['Churn']]
#Entrenando el modelo
model.fit(X_train_o, y_train_o)
#Prediciendo el modelo
y_pred_3 = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred_1)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred_2)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred_3)
```

```python
from sklearn.metrics import classification_report
```

```python
#Original Categoría 1: Precision 0.59 Recall 0.53
print(classification_report(y_test,y_pred_1))
```

```python
#Undersampling Categoría 1: Precision 0.48 Recall 0.76
print(classification_report(y_test,y_pred_2))
```

```python
#Oversampling Categoría 1: Precision 0.49 Recall 0.76
print(classification_report(y_test,y_pred_3))
```
