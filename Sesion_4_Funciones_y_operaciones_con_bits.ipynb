## 1. ¿Qué son operaciones con bit?
Una operación de bit es una función matemática que se aplica a cada bit de un número. Python permite manipular bits de un número entero utilizando operadores de desplazamiento de bits y operadores de enmascaramiento de bits.

Los operadores de desplazamiento de bits son <<, >>. Estos operadores desplazan los bits del número hacia la izquierda (<<) o hacia la derecha (>>) un número específico de posiciones.

````python
# desplazar a la izquierda 3 posiciones
# 5*2^3 = 5*8 = 40
a = 10  # 0101 en binario
b = a << 2
print(b)   #101000 en binario

# 40
# 5*2**3
````

````python
17 * 2**3
````

````python
# desplazar a la derecha 2 posiciones
# 10//2^2 = 10//4 = 2
a = 80   # 1010 en binario
b = a >> 6
print(b)   # 0010 en binario
````

````python
92 // 2**4
````

Los operadores de enmascaramiento de bits son:
* & (AND)
* | (OR)
* ^ (XOR)

Estos operadores permiten realizar operaciones de bits a nivel individual entre dos números.

````python
# operador AND
a = 10     # 0101 en binario
b = 13     # 0011 en binario
c = a & b
print(c)   # 0001 en binario
````

````python
# operador OR
a = 10   # 0101 en binario
b = 13   # 0011 en binario
c = a | b
print(c)   # 0111 en binario
````

````python
# operador XOR
a = 10      # 0101 en binario
b = 13      # 0011 en binario
c = a ^ b
print(c)   # 0110 en binario
````

### 2. Ejercicios prácticos:
Cree una función que devuelva el área y el perímetro de un cuadrado. Solicitar al usuario ingresar 'A' para calcular Area y 'P' para calcular perimetro

* Donde area = lado * lado
* Donde perimetro = lado * 4

````python
#8:45

def f_operacion_area_perimetro(operacion,lado):
  if operacion.upper() == 'A':
    resultado = lado*lado
  elif operacion.upper() == 'P':
    resultado = lado*4
  else:
    resultado = 'No ingreso un tipo de operacion valida'

  return resultado
````

* Area ciruclo = pi * radio **2
* Volumen de un cilindro = Area ciruclo(radio)*altura


````python
#8:50
def f_area_circulo(radio):
  area = 3.1415 * radio **2
  return area
````

````python
def f_volumen_cilindro(radio,altura):
  volumen = f_area_circulo(radio)*altura
  return volumen
````

### 4. Ejercicios prácticos:
Función que convierta grados Celsius a grados Fahrenheit. Fórmula: (C × 9/5) + 32 = F


* Crear un programa que pregunte al usuario si
desea convertir Celsius a Farenheim y se le va a pedir el valor a convertir.
* Alamcenar todas los grados celsius convertidos en una función.
* Imprimir listado de grados de celcius

````python
#8:55
def f_convertir_cel_a_fare(cel):
  fare = (cel * 9/5) + 32
  return fare
````

````python
f_convertir_cel_a_fare(18)
````

````python
lista_grados_cel = list() #creamos lista vacia
respuesta = True # Creamos una variable para crear un bucle infinito

while respuesta:
  pregunta = input('Desea convertir grados celsius a fareinhein (S/N): ')
  if pregunta == 'S':
    g_celsius = int(input('Ingrese los grados en celsius: '))
    lista_grados_cel.append(g_celsius)
    fareinhein = f_convertir_cel_a_fare(g_celsius)
    print('Fareinhein:',fareinhein)
  else:
    respuesta = False
````
